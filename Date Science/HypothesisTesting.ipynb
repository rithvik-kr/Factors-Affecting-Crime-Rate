```python
import pandas as pd
```

```python
df_2009 = pd.read_csv("CrimeRate_2009.csv")
df_2010 = pd.read_csv("CrimeRate_2010.csv")
df_2011 = pd.read_csv("CrimeRate_2011.csv")
```

```python
df_2009.head(100)
```

```python
df_2009.dropna(inplace=True)
df_2010.dropna(inplace=True)
df_2011.dropna(inplace=True)
```

```python
df_2009.loc[0, ["STATE/UT"]] = df_2010.iloc[0]["STATE/UT"]
```

```python
df_police = pd.read_csv("PoliceStrength.csv")
df_police["State"] = df_police["State"].str.upper()
df_police.rename(columns={"State":"STATE/UT"}, inplace=True)
df_police.head(40)
```

```python
result_2009 = pd.merge(df_2009, df_police, how='inner', on='STATE/UT')
result_2010 = pd.merge(df_2010, df_police, how='inner', on='STATE/UT')
result_2011 = pd.merge(df_2011, df_police, how='inner', on='STATE/UT')
result_2011.head(40)
```

```python
from scipy import stats
from scipy.stats import chi2
def hypothesistesting(df, column1, column2):
    contingency_table=pd.crosstab(df[column1],df[column2])
    print('contingency_table :-\n',contingency_table)#Observed Values
    Observed_Values = contingency_table.values 
    print("Observed Values :-\n",Observed_Values)
    b=stats.chi2_contingency(contingency_table)
    Expected_Values = b[3]
    print("Expected Values :-\n",Expected_Values)
    no_of_rows=len(contingency_table.iloc[0:2,0])
    no_of_columns=len(contingency_table.iloc[0,0:2])
    ddof=(no_of_rows-1)*(no_of_columns-1)
    print("Degree of Freedom:-",ddof)
    alpha = 0.05
    chi_square=sum([(o-e)**2./e for o,e in zip(Observed_Values,Expected_Values)])
    chi_square_statistic=chi_square[0]+chi_square[1]
    print("chi-square statistic:-",chi_square_statistic)
    critical_value=chi2.ppf(q=1-alpha,df=ddof)
    print('critical_value:',critical_value)#p-value
    p_value=1-chi2.cdf(x=chi_square_statistic,df=ddof)
    print('p-value:',p_value)
    print('Significance level: ',alpha)
    print('Degree of Freedom: ',ddof)
    print('chi-square statistic:',chi_square_statistic)
    print('critical_value:',critical_value)
    print('p-value:',p_value)
    if chi_square_statistic>=critical_value:
        print("Reject H0,There is a relationship between 2 categorical variables")
    else:
        print("Retain H0,There is no relationship between 2 categorical variables")
        
    if p_value<=alpha:
        print("Reject H0,There is a relationship between 2 categorical variables")
    else:
        print("Retain H0,There is no relationship between 2 categorical variables")
```

```python
hypothesistesting(result_2009, "C1", "2009")
```

```python
hypothesistesting(result_2010, "C1", "2010")
```

```python
hypothesistesting(result_2011, "C1", "2011")
```
